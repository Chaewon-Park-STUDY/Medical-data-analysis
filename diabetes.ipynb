```python
def showPie(data, number):
  data_true=(data> number).value_counts(normalize=True)[True]
  data_false=(data> number).value_counts(normalize=True)[False]
  data_set=[data_true,data_false]
  categories=["True","False"]
  plt.pie(data_set,labels=categories,autopct="%0.1f")
```

```python
showPie(df["Pregnancies"],5)
```

```python
showPie(df["SkinThickness"],30)
```

```python
X=df.drop('Outcome',axis=1)    #Outcome ì—´ ì‚­ì œ
```

```python
X
```

```python
y=df.Outcome
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.20, random_state=0)
```

```python
X
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
LRModel=LogisticRegression(solver='liblinear',random_state=0)
```

```python
LRModel.fit(X_train, y_train)
```

```python
LRModel.score(X_test,y_test)
```

```python
confusion_matrix(y_test,LRModel.predict(X_test))
```

```python
from xgboost import XGBClassifier
XGBModel= XGBClassifier()
XGBModel.fit(X_train,y_train)
```

```python
XGBModel.score(X_test,y_test)
```

```python
confusion_matrix(y_test,XGBModel.predict(X_test))
```

```python
XGBModel.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
RandomForestModel=RandomForestClassifier(n_estimators=100)
RandomForestModel.fit(X_train,y_train)
```

```python
RandomForestModel.score(X_test,y_test)
```

```python
confusion_matrix(y_test,RandomForestModel.predict(X_test))
```
