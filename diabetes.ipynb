```python
# 시각화 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax=plt.subplots(1,1,figsize=(4,4))
sns.distplot(df["Age"])
```

```python
fig, ax=plt.subplots(1,1,figsize=(4,4))
sns.distplot(df["Outcome"])
```

```python
fig, ax=plt.subplots(1,1,figsize=(4,4))
sns.distplot(df["Pregnancies"])
```

```python
fig, ax=plt.subplots(4,2,figsize=(16,16))
sns.distplot(df["Age"], ax=ax[0,0])
sns.distplot(df["Pregnancies"], ax=ax[0,1])
sns.distplot(df["Glucose"], ax=ax[1,0])
sns.distplot(df["BloodPressure"], ax=ax[1,1])
sns.distplot(df["SkinThickness"], ax=ax[2,0])
sns.distplot(df["Insulin"], ax=ax[2,1])
sns.distplot(df["DiabetesPedigreeFunction"], ax=ax[3,0])
sns.distplot(df["BMI"], ax=ax[3,1])
```
